# Data Cleaning with basic pandas functions

 - We are cleaning the data using basic pandas function to have a clear view
 - we are not identifying outliers in this project


```python
#import pandas module to read the csv file in dataframe format
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/shakt/OneDrive/Desktop/CSV files/SAMPLE.csv")
df
```

```python
#dropping the rows which have all the column values as null
df.dropna(how='all',inplace=True)
df
```

```python
#'sl No' should be unique so we are finding if any duplicates are there in the column. If so, we need to keep one and remove remaining.
df[df['sl No'].duplicated(keep=False)]
```

```python
# we have found three duplicates with 'sl No' column so removing two by referring to the index values of the rows.(in this example 19th index has all the details )
df=df.drop([3,10])
df
```

```python
#checking if still null values are there in the dataframe
df.isna().sum()
```

```python
# removing rows if any null values are there as we cannot substitute with any values all are unique in their own way
df.dropna(how='any', inplace=True)
df
```

```python
# finding one college name is odd so changing the name assuming it should be 'DBC' but in real we need to confirm with the client
df['college'].replace({'dhgvuicbh':'DBC'},inplace=True)
df
```

```python
#sorting the dataframe based on name column in ascending order
df.sort_values('name', ascending=True)
```
